```python
import csv 
import matplotlib.pyplot as plt #plt is short hand for matplotlib
import numpy as np
t=[]
data=[]
path = "data/"
with open (path + '10_clean_pulses.csv') as f: 
    reader = csv.reader(f)
    for row in reader: 
        t.append(float(row[0]))
        data.append(float(row[1]))
```

```python
step = 10
stop = len(data)

x = t[0: stop: step]
y = data[0: stop: step]
```

```python
t[stop-1]
```

```python
plt.plot(x, y)

y_len = len(y)
interval = t[stop-1] /10 # length of an interval, 10 is number of heartbeats
iterations = 1600 / step # iterations per second
interval_len = interval * iterations
interval_len = int(interval_len)

index = 0
while (index + interval_len <= y_len):
    endIndex = index + interval_len
    minValue = min(y[index:endIndex])
    
    minIndex = y.index(minValue, index, endIndex)
    
    plt.scatter(x[minIndex], minValue, color = "r", marker = "*")
    index = endIndex

plt.title("Ten Heart Beat Pulses")
plt.xlabel("time (seconds)")
plt.ylabel("voltage")

plt.show()
```
