```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Plotting
- ranges
- setting tickmarks
- plotting mins
- title, xlabel, ylabel

```python
end = 10 * math.pi

x = np.arange(0, end, math.pi/ 8)
y = np.sin(x)
plt.plot(x, y)

# find min interval math
x_len = len(x)
interval = 2 * math.pi # interval time
interval_count = end / interval # number of intervals
interval_len = x_len/ interval_count # iterations per interval
interval_len = int(interval_len) # convert to int

index = 0
while (index + interval_len <= x_len):
    endIndex = index + interval_len
    minValue = min(y[index:endIndex])
    minIndex = np.where(y[index: endIndex] == minValue) 
    plt.scatter(x[minIndex[0] + index], minValue, color = "r")
    index = endIndex
# set title and x label
plt.title("this is the title")
plt.xlabel("x")
plt.ylabel("y")

# creating tick ranges
x_ticks = np.arange(math.pi/2, end, 2 * math.pi)
my_xticks = ['pi/2','5pi/2', '9pi/2', '13pi/2', '17pi/2', "21pi/2"]
y_ticks = np.arange(-1, 1, math.pi/8)

plt.xticks(x_ticks, my_xticks)
plt.yticks(y_ticks)

plt.show()
```

# Subplots

```python
end = 10 * math.pi

x = np.arange(0, end, math.pi/ 8)
y = np.sin(x)
y2 = np.sin(x)

fig, axes = plt.subplots(2, 1)
axes[0].plot(x, y)
axes[1].plot(x, y2)

axes[0].set_title("plot 1")
axes[1].set_title("plot 2")

axes[0].set_ylabel("y1")
axes[1].set_ylabel("y2")

axes[0].set_xlabel("x1")
axes[1].set_xlabel("x2")

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

left  = 0.125  # the left side of the subplots of the figure
right = 0.9    # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.2   # the amount of width reserved for blank space between subplots
hspace = 0.2   # the amount of height reserved for white space between subplots

plt.show()
```
